```
def parse_txt(txt_file):

    with open(txt_file, 'r') as f :

        lines = f.readlines()
        dict = {}

        for line in lines :
            splits = line.split()
            if(len(splits)!=3) : 
                continue
            splits = list(map(lambda x : x.strip(), splits))
            if splits[0] in dict and splits[2] < dict[splits[0]][1] :
                continue
            dict[splits[0]] = (splits[1],splits[2])

        res = list( map(lambda x : (dict[x][0], x), list(dict.keys()) ))
        return  res
```

```
import xml.etree.ElementTree as ET

def generate_xml(root_xml_file, names_pairs, output_file_name = './output.xml') :

    doc_tree = ET.parse(root_xml_file)
    doc_root = doc_tree.getroot()
    id = len(doc_root.findall('Name'))+1


    for name_pair in name_pairs :
        name = ET.Element('Name')
        name.attrib['ID'] = str(id)
        source_name = ET.SubElement(name, 'SourceName')
        source_name.text = name_pair[0]
        target_name = ET.SubElement(name, 'TargetName')
        target_name.text = name_pair[1]
        target_name.attrib['ID'] = '1'
        doc_root.append(name)
        id+=1

    doc_tree.write(output_file_name)
```

```
name_pairs = parse_txt('./hi.translit.sampled.train.tsv')
generate_xml('./english-hindi-train.xml', name_pairs)
```
